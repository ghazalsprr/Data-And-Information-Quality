```python
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# select the style from fivethityeight website
plt.style.use('fivethirtyeight') 
mpl.rcParams['lines.linewidth'] = 2

# predefined figsize
figsize=(12,12)

import warnings
warnings.filterwarnings('ignore')



%config InlineBackend.figure_format = 'png'
%matplotlib inline
```

```python
from scipy.stats import pearsonr

def corrfunc(x,y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    r, _ = pearsonr(x, y)
    ax = ax or plt.gca()
    # Unicode for lowercase rho (ρ)
    rho = '\u03C1'
    ax.annotate(f'{rho} = {r:.2f}', xy=(.1, .9), xycoords=ax.transAxes)
    
def corrdot(*args, **kwargs):
    corr_r = args[0].corr(args[1], 'spearman')
    corr_text = f"{corr_r:2.2f}".replace("0.", ".")
    ax = plt.gca()
    ax.set_axis_off()
    marker_size = abs(corr_r) * 10000
    ax.scatter([.5], [.5], marker_size, [corr_r], alpha=0.6, cmap="Blues",
               vmin=-1, vmax=1, transform=ax.transAxes)
    font_size = abs(corr_r) * 40 + 5
    ax.annotate(corr_text, [.5, .5,],  xycoords="axes fraction",
                ha='center', va='center', fontsize=font_size)    
    
# g = sns.pairplot(stocks,palette=["Blues_d"])

def plot_corr(dataset : pd.DataFrame , numerical_columns : list): 
    g = sns.PairGrid(dataset[numerical_columns], aspect=1.4, diag_sharey=False)
    g.map_lower(corrfunc)
    g.map_lower(sns.regplot,  ci=False, line_kws={'color': 'Black','linewidth':1})
    g.map_diag(sns.distplot, kde_kws={'color': 'Black','linewidth':1})
    g.map_upper(corrdot)
    plt.figure(figsize=(20, 20))
    plt.show()


#select at most 15 numerical columns to plot the correlation matrix, more columns will take too long to compute and the plot will be too crowded
plot_corr(dataset, numerical_columns[20:35])
```

# Data Wranling

```python
columnsdrop = ['photoUrl','playerUrl','Loan Date End']
dataset = dataset.drop(columns = columnsdrop)

```

```python
dataset.rename(columns={'LongName':'FullName'},inplace=True)


dataset['Hits']=dataset['Hits'].str.replace(r'\n', '', regex=True)
dataset['Club']=dataset['Club'].str.replace(r'\n\n\n\n', '', regex=True)
# Convert the "↓OVA" column to "OVA" (rename for simplicity)
dataset.rename(columns={"↓OVA": "OVA"}, inplace=True)

```

```python
# Extract unique weight values (before cleaning)
print(dataset['Weight'].unique())

# Use regex to remove 'kg' or 'lbs' from the Weight column
dataset['Weight'] = dataset['Weight'].str.replace(r'(kg|lbs)', '', regex=True)

# Display the first few rows to confirm changes
dataset.head()
```

```python
# Clean financial columns ('Value', 'Wage')
# Convert Value and Wage to numeric values in millions
def convert_money(value):
    value = str(value).replace("€", "").replace("M", "").replace("K", "").strip()
    if "M" in str(value):
        return float(value) * 1_000_000  # Convert millions to full number
    elif "K" in str(value):
        return float(value) * 1_000      # Convert thousands to full number
    else:
        return float(value) if value else 0

dataset["Value"] = dataset["Value"].apply(convert_money)
dataset["Wage"] = dataset["Wage"].apply(convert_money)
dataset["Wage"]
dataset["Value"]
```

# Outliers Detection

```python
from scipy.stats import zscore
import pandas as pd

# Step 1: Define relevant numerical columns for outlier detection
numerical_columns = ["OVA", "POT", "PAC", "SHO", "PAS", "DRI", "DEF", "PHY", "Value", "Wage"]

# Ensure all relevant columns are numeric
dataset[numerical_columns] = dataset[numerical_columns].apply(pd.to_numeric, errors="coerce")

# Drop rows with NaN in numerical columns
dataset = dataset.dropna(subset=numerical_columns)

# Step 2: Z-Score Method
# Compute Z-scores for the numerical columns
z_scores = dataset[numerical_columns].apply(zscore)

# Set threshold for outliers (e.g., |Z| > 3)
z_outliers = (z_scores.abs() > 3).sum()

# Step 3: IQR Method
# Compute Q1, Q3, and IQR for each numerical column
Q1 = dataset[numerical_columns].quantile(0.25)
Q3 = dataset[numerical_columns].quantile(0.75)
IQR = Q3 - Q1

# Detect outliers using the IQR method
iqr_outliers = ((dataset[numerical_columns] < (Q1 - 1.5 * IQR)) |
                (dataset[numerical_columns] > (Q3 + 1.5 * IQR))).sum()

# Combine results into a summary
outlier_summary = pd.DataFrame({
    "Z-Score Outliers": z_outliers,
    "IQR Outliers": iqr_outliers
})

# Display the summary
print(outlier_summary)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
 
# Prepare data for visualization 
outlier_summary['Feature'] = outlier_summary.index 
 
# Melt the data for easier plotting 
outlier_data_melted = outlier_summary.melt(id_vars='Feature', var_name='Method', value_name='Outlier Count') 
 
# Plot the outlier counts using a bar plot 
plt.figure(figsize=(12, 6)) 
sns.barplot(data=outlier_data_melted, x='Feature', y='Outlier Count', hue='Method') 
 
# Customize the plot 
plt.title("Outlier Counts by Feature and Detection Method", fontsize=16) 
plt.xlabel("Feature", fontsize=14) 
plt.ylabel("Outlier Count", fontsize=14) 
plt.xticks(rotation=45, fontsize=12) 
plt.legend(title="Detection Method", fontsize=12) 
plt.tight_layout() 
 
# Show the plot 
plt.show()
```
